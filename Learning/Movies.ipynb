```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'D:\DS\DataSets\movies.csv')
```

```python
df.shape
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
mean_rating =  df.RATING.mean()
round(mean_rating)
```

```python
#  Filling the average rating for NaN values
df.fillna(value = {'RATING':round(mean_rating)}, inplace = True)
df.fillna(value = {'Gross': '0'}, inplace = True)
```

```python
df.head()
```

```python
df['Gross'] = df['Gross'].apply(lambda x : x[1:-1] if x != '0' else '0')
df.Gross = df['Gross'].astype(float)
```

```python
df.Gross.replace( {0.0 : (df.Gross.mean())}, inplace = True)
df.Gross = round(df['Gross'])
df.Gross.astype(int)
df.head()
```

```python
df.fillna(value = {'RunTime': (df['RunTime'].mean())}, inplace = True)
df['RunTime'] = round(df['RunTime'])
df['RunTime'].astype(int)
df.head()
```

```python
def CommaRemove(vote):
    lst = vote.split(',')
    return ''.join(lst)
```

```python
df.fillna(value = {'VOTES' : '0'}, inplace = True)
df['VOTES'] = df['VOTES'].apply(lambda x : CommaRemove(x))
```

```python
df['VOTES'] = df['VOTES'].astype('int')
```

```python
df.info()
```

```python
df.head(5)
```

```python
def newLineSep(content):
    tmp = content.split('\n')
    tmp1 = ''.join(tmp)
    tmp2 = tmp1.split('|    ')
    return ''.join(tmp2)
```

```python
df['STARS']
```

```python
df['STARS'] = df['STARS'].apply(lambda x : newLineSep(str(x)))
df['ONE-LINE'] = df['ONE-LINE'].apply(lambda x : newLineSep(str(x)))
df['GENRE'] = df['GENRE'].apply(lambda x : newLineSep(str(x)))
df.head(5)
```

```python
def Year(year):
    tmp = []
    dig = '0123456789'
    for i in year:
        if i in dig:
            tmp.append(i)
    tmp1 = ''.join(tmp)
    if len(tmp1) > 4:
        return tmp1[:4] + ';' + tmp1[4:]
    else:
        return tmp1
```

```python
df['YEAR'] = df['YEAR'].apply(lambda x : Year(str(x)))
df.head(5)
```

```python
df.columns = map(str.capitalize,df.columns)
df.head()
```

```python
df['Gross'] = df['Gross'].apply(lambda x : x*1000000)
df.head()
```

```python
df.to_csv('D:\DS\DataSets\movies_cleaned.csv',index = False)
```
