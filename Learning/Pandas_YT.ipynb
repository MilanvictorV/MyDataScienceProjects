```python
data = {
    'first': ['Milan','Linssie','Nithish'],
    'last' : ['victor','T','Shanmugam'],
    'email':['mv@abc.com','lt@abc.com','ns@abc.com']
}
```

```python
# to access the rows randomly
# iloc full ah vae only integers so we must give index values
#syntax iloc[[row_indexes],column_indexes]
df.iloc[0] # indexing
df.iloc[[0,1],2] # we are taking email id column for first 2 rows
df.iloc[[0,1],[0,1]] # we can give index for columns also
```

```python
# loc same as iloc in here for column we must give columns names only.
df.loc[[1,2],['email','first']]
```

```python
df.head(2)
```

```python
# Filtering
# We are going to see some simple filtering
# We are getting hobbyist as No and will do some changes
df['Hobbyist'] == 'Yes' # it will return true if it's matching or false
df[df['Hobbyist'] == 'Yes']
# using loc we can get the required columns
filt = (df['Hobbyist'] == 'Yes')
df.loc[filt,'MainBranch']
```

```python
# we are going to add some other conditions
#df.head(2)
filt = ((df['Hobbyist'] == 'No') & (df['OpenSourcer'] == 'Never'))
df[filt]
df[~filt]# it will give opposite of filt query
```

```python
# isin
cty = ['India','Canada']
filt = df['Country'].isin(cty)
df[filt]
```

```python
# to find a word in an string
filt = df['LanguageWorkedWith'].str.contains('Python',na = False) # we are not considering the NaN values so na = False is used
df.loc[filt,'LanguageWorkedWith']
```

```python
# Updating our DataFrame and columns
# renaming the columns
# df.columns = ['','',..] ella column kum name kudukanum here we have more than 85 columns so not renaming it
# upper casing the column name
df.columns = [x.upper() for x in df.columns]
df.columns # it's a permanant changes of column name
```

```python
# we are going to test replace method which is available in str class
#df.columns.str.replace('#','A')
df.columns.str.replace(' ','_') # replacing the space by underscore
```
