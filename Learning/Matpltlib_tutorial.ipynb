# Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = [1,2,3,4,5]
b = [23,4,13,22,41]
plt.xlabel('X')
plt.ylabel('Y')
plt.title('SAMPLE')
plt.plot(a,b,color = 'green',linewidth = 3.5,linestyle = 'dotted')
plt.show()
```

```python
#another optimized way
x1 = [1,2,3,4,5,6,7]
y1 = [50,51,54,23,56,54,35]
x2 = [1,2,3,4,5,6,7]
y2 = [55,54,54,63,46,34,25]
```

```python
plt.plot(x2,y2,color = '#111111',marker = 'o',linestyle = '--',markersize = 10)
```

```python
plt.plot(x1,y1,'g+-.', x2,y2,'rD--',linewidth = 3,alpha = 0.6) #green color with marker as + by changing the linestyle
#alpha determines the trans[arancy]
```

```python
#plotting multiple charts in one graph
days = [1,2,3,4,5,6,7]
maxt = [35,34,32,40,42,45,47] #maximum temperature
mint = [30,29,27,25,10,14,35] #minimum temperatur
avgt = [32,33,30,36,25,28,40] #average temperature
```

```python
plt.title('Weather')
plt.xlabel('Temperature')
plt.ylabel('days')
plt.plot(days,maxt,'g+--',alpha = 0.5,label = 'Maximum Temp')
plt.plot(days,mint,'ro-.',alpha = 0.3,label = 'Minimum Temp')
plt.plot(days,avgt,'bD',linestyle = 'dotted',alpha = 0.8,label = 'Average Temp')
plt.legend(loc = 'lower left',shadow = True)
plt.grid()
plt.show()
```

## Bar chart

```python
comp = ['Google','Microsoft','Apple','Facebook']
rev = [150,70,100,75]
pro = [45,40,35,43]
```

```python
plt.title('Revenue 2022')
plt.xlabel('Companies')
plt.ylabel('Revenue(Bn)')
plt.bar(comp,rev,label = 'Revenue')
plt.bar(comp,pro,label = 'Profit')
#plt.grid()
plt.legend()
plt.show()
```

```python
# to display the above chart as side by side chart
xpos = np.arange(len(comp))
xpos
```

```python
plt.title('Revenue 2022')
plt.xlabel('Companies')
plt.ylabel('Revenue(Bn)')
plt.xticks(xpos,comp)
plt.bar(xpos-0.2,rev,label = 'Revenue',width = 0.4)
plt.bar(xpos+0.2,pro,label = 'Profit',width = 0.4)
#plt.grid()
plt.legend()
plt.show()
```

```python
#to display it in Horizondal chart
plt.title('Revenue 2022')
plt.ylabel('Companies')
plt.xlabel('Revenue(Bn)')
plt.yticks(xpos,comp)
plt.barh(xpos-0.2,rev,label = 'Revenue',height =0.4)
plt.barh(xpos+0.2,pro,label = 'Profit',height =0.4)
#plt.grid()
plt.legend()
plt.show()
```

## Histogram

```python
bsm = np.random.randint(80,180,50)
bsf = np.random.randint(80,150,50)
```

```python
#Histogram works on one dimensional array
#By default bin size is 10
# plt.hist(bsm,bins = (80,120,160,200),rwidth = 0.95,color = 'g',histtype = 'step')
# plt.hist(bsf,bins = (80,120,160,200),rwidth = 0.95,color = 'r',histtype = 'step')
plt.title('Blood Sugar Level',color = 'green',fontsize = 15)
plt.xlabel('Sugar Range',color = 'Red')
plt.ylabel('No of Persons',color = 'blue')
plt.hist([bsm,bsf],bins = (80,120,160,200),rwidth = 0.95,color = ['r','orange'],label=['Blood sugar level - Men','Blood sugar level - Women'])
plt.legend()
#rwidth - relational width of each bin, another bins = we can give range(80,100,150,200)
plt.show()
```

```python
#Histogram works on one dimensional array
#By default bin size is 10
# plt.hist(bsm,bins = (80,120,160,200),rwidth = 0.95,color = 'g',histtype = 'step')
# plt.hist(bsf,bins = (80,120,160,200),rwidth = 0.95,color = 'r',histtype = 'step')
plt.title('Blood Sugar Level',color = 'green',fontsize = 15)
plt.xlabel('No of Persons',color = 'Red')
plt.ylabel('Sugar Range',color = 'blue')
plt.hist([bsm,bsf],bins = (80,120,160,200),rwidth = 0.95,color = ['r','orange'],label=['Blood sugar level - Men','Blood sugar level - Women'],
        orientation='horizontal')
plt.legend()
#rwidth - relational width of each bin, another bins = we can give range(80,100,150,200)
plt.show()
```

# Pie Chart

```python
exp = [6000,3000,1000,1500,10000,2500]
lab = ['Rent','EMI','Petrol','Internet&EB','Savings','Others']
```

```python
plt.pie(exp,labels = lab,
        shadow = True,
        autopct = '%0.2f%%',
        radius = 1.2,
        startangle = 180,
        explode=[0,0,0.5,0.5,0,0])
#explode vandhu endha edathula pie velia varanumo atha mention pannum evlo pie value iruko avlo array value pass pannanum
#plt.axis('equal')
plt.title('My Monthly Expense',color = 'g',fontsize = 13)
plt.show()
```
