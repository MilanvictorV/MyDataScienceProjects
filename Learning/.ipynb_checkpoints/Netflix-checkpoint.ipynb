```python
# To Rearrange the duration, we are considering 8 episodes per season and each episode is 30 minutes
def DurationChange(Duration):
    tmp = Duration.split(' ')
    if tmp[1] == 'min' or tmp[1] == 'Min':
        return tmp[0]
    elif tmp[1] == 'Seasons' or tmp[1] == 'Season':
        return int(tmp[0]) * 8 * 30
```

```python
df.Duration = df['Duration'].apply(lambda x : DurationChange(x))
df.Duration = df.Duration.astype('float')
df['Duration'] = df['Duration'].apply(lambda x : "{:.1f}".format(x/60))
```
