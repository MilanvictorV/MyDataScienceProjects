```python
import pandas as pd
```

```python
data1 = {
    'ID' : ['1','2','3','4','5'],
    'Feature1' : ['A','B','C','D','E'],
    'Feature2' : ['L','M','N','O','P']
}
```

```python
 df1 = pd.DataFrame(data1,columns = ["ID","Feature1","Feature2"])
```

```python
df1
```

```python
d2 = {
    'ID' : ['1','3','6','7','8'],
    'Feature1' : ['AB','BC','CD','DE','EF'],
    'Feature2' : ['KL','LM','M-N','N-O','O-P']
}
```

```python
df2 = pd.DataFrame(d2,columns = ['ID','Feature1','Feature2'])
#df3 = pd.DataFrame(d2,columns = ['ID1','Feature11','Feature22'])
df2
```

```python
#By concatinating 2 different columns
#DF = pd.concat([df1,df2])
DF = pd.concat([df1,df2])
DF
```

```python
#to arrange the index properly
DF = pd.concat([df1,df2],ignore_index = True)
DF
```

```python
DF = pd.concat([df1,df2],keys = ['1st','2nd'])
DF
```

```python
DF = pd.concat([df1,df2],sort = True) #if we don't give any column it will sort by column names
DF
```

```python
DF['Feature3'] = DF['Feature1'] + DF['Feature2']
DF
```

```python
merge_i = pd.merge(df1,df2, on = 'ID', how  = 'inner')
merge_o = pd.merge(df1,df2, on = 'ID', how  = 'outer')
merge_r = pd.merge(df1,df2, on = 'ID', how  = 'right')
merge_l = pd.merge(df1,df2, on = 'ID', how  = 'left')
merge_l
```
