```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1 = pd.read_csv(r'https://raw.githubusercontent.com/MilanvictorV/MyDataScienceProjects/main/Data_Analysis_projects/Credit_Card_Approval/application_record.csv')
df2 = pd.read_csv(r'https://raw.githubusercontent.com/MilanvictorV/MyDataScienceProjects/main/Data_Analysis_projects/Credit_Card_Approval/credit_record.csv')
```

```python
df1.head()
```

```python
df2.head()
```

```python
df = pd.merge(left = df1, right = df2, how = 'inner')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
#Hereafter no need for ID column so dropping the column
df.drop(columns = 'ID', inplace = True)
df.head()
```

```python
df.columns = df.columns.str.lower()
df.head()
```

```python
df.isna().sum()
```

```python
df['occupation_type'].mode()
```

```python
#Laborers
df.fillna({'occupation_type' : 'Laborers'}, inplace = True)
df.isna().sum()
```

```python
df.head()
```

```python
df['code_gender'].unique()
```

```python
# Assigning 0 for Male and 1 for Female
df['code_gender'] = df['code_gender'].apply(lambda x : 0 if x == 'M' else '1')
df.head()
```

```python
df['flag_own_car'].unique()
```

```python
# Assigning 0 if No car and 1 if own a car
df['flag_own_car'] = df['flag_own_car'].apply(lambda x : 0 if x == 'N' else '1')
df.head()
```

```python
df['flag_own_realty'].unique()
```

```python
#  Doing it same as Car
df['flag_own_realty'] = df['flag_own_realty'].apply(lambda x : 0 if x == 'N' else '1')
df.head()
```

```python
df['name_income_type'].unique()
```

```python
# Now Converting Gender, Car, Realty columns to int
df['code_gender'] = df['code_gender'].astype(int)
df['flag_own_car'] = df['flag_own_car'].astype(int)
df['flag_own_realty'] = df['flag_own_realty'].astype(int)
df.info()
```

```python
# Function to find the days to date converter
def dateconvert(date):
    start_date = datetime.today()
    days_birth = -12005
    birth_date = start_date + timedelta(days=days_birth)
    return birth_date.strftime("%Y-%m-%d")
```

```python
df['age'] = round((-1 * df['days_birth']/365))
df['work_exp'] = round((-1 * df['days_employed']/365))
df.drop(columns = ['days_birth','days_employed'] , inplace = True)
df.head()
```
